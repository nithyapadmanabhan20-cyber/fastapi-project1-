```python

from fastapi import FastAPI

app = FastAPI()

users = []
tasks = []


# REGISTER
@app.post("/register")
def register(username: str, password: str):
    users.append({
        "username": username,
        "password": password
    })

    return {"message": "User created"}


# LOGIN
@app.post("/login")
def login(username: str, password: str):

    for user in users:
        if user["username"] == username and user["password"] == password:
            return {"message": "Login successful"}

    return {"message": "Invalid details"}


# CREATE TASK
@app.post("/tasks")
def create_task(title: str):

    tasks.append({
        "title": title,
        "completed": False
    })

    return {"message": "Task added"}


# VIEW TASKS
@app.get("/tasks")
def get_tasks():
    return tasks


# COMPLETE TASK
@app.put("/tasks")
def complete_task(title: str):

    for task in tasks:
        if task["title"] == title:
            task["completed"] = True
            return {"message": "Task completed"}

    return {"message": "Task not found"}


# DELETE TASK
@app.delete("/tasks")
def delete_task(title: str):

    for task in tasks:
        if task["title"] == title:
            tasks.remove(task)
            return {"message": "Task deleted"}

    return {"message": "Task not found"}
```
